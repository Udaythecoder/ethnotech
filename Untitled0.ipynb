```python
import matplotlib.pyplot as plt

months = ['Jan','Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [10, 15, 13, 18, 22, 25]

plt.figure(figsize=(8, 5))
plt.plot(months, sales, marker='o', color='b', linestyle='-', linewidth=2)

plt.title('Monthly Sales Trend', fontsize=14, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Sales (in $1,000s)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
products = ['Apples', 'Bananas', 'Oranges', 'Berries']
quantities = [55, 85, 43, 67]
colors = ['red', 'gold', 'orange', 'purple']

# Create the plot
plt.figure(figsize=(8, 5))
plt.bar(products, quantities, color=colors, edgecolor='black', alpha=0.8)

# Customization
plt.title('Fruit Stock Inventory', fontsize=14, fontweight='bold')
plt.xlabel('Fruit Type', fontsize=12)
plt.ylabel('Quantity in Stock', fontsize=12)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating random data (e.g., heights of 200 people)
np.random.seed(42)
heights = np.random.normal(loc=170, scale=10, size=200)

# Create the plot
plt.figure(figsize=(8, 5))
plt.hist(heights, bins=15, color='teal', edgecolor='black', alpha=0.7)

# Customization
plt.title('Distribution of Heights', fontsize=14, fontweight='bold')
plt.xlabel('Height (cm)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
study_hours = [1, 2, 3, 4, 5, 5.5, 6, 7, 8, 9, 10]
exam_scores = [45, 50, 55, 60, 65, 62, 70, 78, 85, 88, 95]

# Create the plot
plt.figure(figsize=(8, 5))
plt.scatter(study_hours, exam_scores, color='darkorange', s=100, edgecolor='black', alpha=0.8)

# Customization
plt.title('Study Hours vs. Exam Scores', fontsize=14, fontweight='bold')
plt.xlabel('Hours Studied', fontsize=12)
plt.ylabel('Exam Score (%)', fontsize=12)
plt.grid(True, linestyle=':', alpha=0.6)

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Setting a clean aesthetic style
sns.set_theme(style="whitegrid")

# Using a built-in Seaborn dataset (Tips in a restaurant)
tips = sns.load_dataset("tips")

# Create the plot
plt.figure(figsize=(8, 5))
sns.boxplot(x="day", y="total_bill", data=tips, palette="Set2")

# Customization
plt.title('Total Bill Distribution by Day', fontsize=14, fontweight='bold')
plt.xlabel('Day of the Week', fontsize=12)
plt.ylabel('Total Bill ($)', fontsize=12)

# Show the plot
plt.show()
```
