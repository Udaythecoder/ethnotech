```python
import pandas as pd
import numpy as np
data={
    "name":["uday","nandha","santhu","vignesh","vicky","sai"],
    "N_type":["jio","airtel","bsnl","jio","airtel","jio"],
    "N_speed(kb/s)":[220,230,200,245,235,240],
    "c_service":["good","good","bad","average","average","good"],
    "user_id": [101, 102, 103, 104, 105, 106],
    "age": [25, 30, 45, 22, 35, 28],
    "monthly_bill_amount": [302, 312, 250, 340,349,299],
    "churned_in_last_year": [1, 2, 4, 0, 0, 0]
}

df = pd.DataFrame(data)
print(df)
```

```python
avg_bill_by_service = df.groupby('c_service')['monthly_bill_amount'].mean().reset_index()
display(avg_bill_by_service)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

churn_by_N_type = df.groupby('N_type')['churned_in_last_year'].sum().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(x='N_type', y='churned_in_last_year', data=churn_by_N_type, palette='viridis')
plt.title('Churned Users Count by N_type')
plt.xlabel('Network Type')
plt.ylabel('Number of Churned Users')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
