```python
import pandas as pd
import numpy as np
data={
    "name":["uday","nandha","santhu","vignesh","vicky","sai"],
    "N_type":["jio","airtel","bsnl","jio","airtel","jio"],
    "N_speed(kb/s)":[220,230,200,245,235,240],
    "c_service":["good","good","bad","average","average","good"],
    "user_id": [101, 102, 103, 104, 105, 106],
    "age": [25, 30, 45, 22, 35, 28],
    "monthly_bill_amount": [302, 312, 250, 340,349,299],
    "churned_in_last_year": [1, 2, 4, 0, 0, 0]
}

df = pd.DataFrame(data)
print(df)
```

```python
avg_bill_by_service = df.groupby('c_service')['monthly_bill_amount'].mean().reset_index()
display(avg_bill_by_service)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

churn_by_N_type = df.groupby('N_type')['churned_in_last_year'].sum().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(x='N_type', y='churned_in_last_year', data=churn_by_N_type, palette='viridis')
plt.title('Churned Users Count by N_type')
plt.xlabel('Network Type')
plt.ylabel('Number of Churned Users')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd

water_data = {"days":[1,2,3,4,5,6,7],
              "waterc":[200,210,150,240,180,220,270]}
for i in range (len(water_data["days"])):
  print(water_data["days"][i],water_data["waterc"][i])
```

```python
water_df = pd.DataFrame(water_data)
total_weekly_water_consumption = water_df['waterc'].sum()
print(f"Total weekly water consumption: {total_weekly_water_consumption} units")
```

```python
import numpy as np
print(np.average(water_data['waterc']))
```

```python
st_avg=water_df['waterc'].std()
print(st_avg)
```

```python
waterc_diff=water_df['waterc'].diff()
print(waterc_diff)
```

```python
import numpy as np
avg = np.average(waterc_diff.dropna())
print(avg)
```
