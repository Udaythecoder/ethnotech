```python
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv('/content/healthcare_dataset.csv')
data.dropna(inplace=True)
data.drop_duplicates(inplace=True)

plt.figure(figsize=(15,6))
data1=data.head(7)
plt.bar(data1['Name'],data1['Billing Amount'])
plt.xlabel("Name")
plt.ylabel("Billing_Amount")
plt.title("healthcare analysis")
plt.show()
```

```python
blood_type_counts = data['Blood Type'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(blood_type_counts, labels=blood_type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Blood Types')
plt.axis('equal')
plt.show()
```

```python
d2=data.head(20)
plt.figure(figsize=(10, 6))
plt.scatter(d2['Age'], d2['Billing Amount'], alpha=0.6)
plt.xlabel('Age')
plt.ylabel('Billing Amount')
plt.title('Age vs. Billing Amount (Scatter Plot)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
import seaborn as sns
d2=data.head(20)
plt.figure(figsize=(14, 7))
sns.boxplot(x='Medical Condition', y='Billing Amount', data=d2)
plt.xlabel('Medical Condition')
plt.ylabel('Billing Amount')
plt.title('MC VS BA')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

#data2=data.head(10)
treemap_data_plotly = data.groupby(['Medical Condition', 'Gender'])['Billing Amount'].sum().reset_index()

fig = px.treemap(treemap_data_plotly,
                 path=[px.Constant("All"), 'Medical Condition', 'Gender'],
                 values='Billing Amount',
                 title='Billing Amount by Medical Condition and Gender (Plotly Treemap)')

fig.update_layout(margin = dict(t=50, l=25, r=25, b=25))
fig.show()
```

```python
import seaborn as sns

data2=data.head(10)
heatmap_data = data2.groupby(['Medical Condition', 'Gender'])['Billing Amount'].mean().unstack()

plt.figure(figsize=(14, 8))
sns.heatmap(heatmap_data, annot=True, fmt=".0f", cmap="viridis", linewidths=.5)
plt.title('Average Billing Amount by Medical Condition and Gender')
plt.xlabel('Gender')
plt.ylabel('Medical Condition')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

gender_medical_condition_counts = data.groupby(['Medical Condition', 'Gender']).size().reset_index(name='Count')
plt.figure(figsize=(10, 10))
plt.pie(gender_medical_condition_counts['Count'],
        labels=gender_medical_condition_counts.apply(lambda x: f"{x['Medical Condition']} - {x['Gender']}", axis=1),
        autopct='%1.1f%%',
        startangle=90)
plt.title('Distribution of Medical Condition and Gender')
plt.axis('equal')
plt.show()
```
