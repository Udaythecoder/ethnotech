```python
import statistics
people=["uday","vicky","vignesh","santhu","sai"]
age=[21,18,20,19,21]
whr=[4,4.5,6,3,7]
clrs=[225,250,300,150,400]
print("watch hour dataset")
for i in range(len(people)):
  print(people[i],age[i],whr[i],clrs[i])

average_whr=sum(whr)/len(whr)
print("average watch hour",average_whr)
```

```python
maximum_clrs=max(clrs)
print("maximum calories",maximum_clrs)

minimum_clrs=min(clrs)
print("minimum calorties",minimum_clrs)

median_clrs=statistics.median(clrs)
print("median calories",median_clrs)

mode_clrs=statistics.mode(clrs)
print("mode calories",mode_clrs)
```

```python
import statistics
std_clrs=statistics.stdev(clrs)
print("standard deviatiation calories",std_clrs)

index=clrs.index(maximum_clrs)
print("most active people:",people[index])

index=clrs.index(minimum_clrs)
print("least active people:",people[index])
```
